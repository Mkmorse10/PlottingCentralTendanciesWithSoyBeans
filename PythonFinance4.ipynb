```python
df['CO'] = (df['Close'] - df['Open'])/df['Open']
df['Median']= (df['High'] + df['Low'])/2
df['Midpoint'] = (df['High'] - df['Low'])/2
df['Log']= np.log(df['Close']/df['Close'].shift())
df['rs'] = (df['Log']-df['Log'].mean()/df['Log'].std())
df = pd.DataFrame(df)
df.dropna(inplace=True)
```
